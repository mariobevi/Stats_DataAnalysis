**3) GDP per Capita**

H0: μQ1 - μRest ≥ 0

H1: μQ1 - μRest < 0

```
my_data.gdp_pc.describe()

q1 = my_data[my_data.gdp_pc <= 0.6595]
rest_q = my_data[my_data.gdp_pc > 0.6595]
n=153
alfa=0.05
conf=95
alfa_half=0.025
df= n-1 #or smallest sample size group - 1


#HO: μq1 - μrest => 0
#H1: μq1 - μrest < 0
print ("HO: μq1 - μrest => 0 ")
print ("H1: μq1 - μrest < 0 0")
print ("This is an lower tail test")

#Q1 data
x_bar1= q1.happ_score.mean()
std1= q1.happ_score.std()
n1= q1.happ_score.count()
print (x_bar1, std1, n1)

#Rest Data
x_bar2= rest_q.happ_score.mean()
std2= rest_q.happ_score.std()
n2= rest_q.happ_score.count()
print (x_bar2, std2, n2)

#CRITICAL POINTS APPROACH:

tscore1=(stats.t.ppf(alfa,df)) #1-alfa since its only an upper tail test, if it was a 2 tail, put 1-alfa_half
print ("Tc1:", round (tscore1,3))

#Compute T empirical
Te= (x_bar1-x_bar2)/(math.sqrt((std1**2/n1)+(std2**2/n2)))
print ("Te:", Te)


#P_VALUE APPROACH:

prob_up_te= stats.t.cdf(abs(Te), df)

p_val= (1-prob_up_te)
print ("p_val:", round(p_val,3))
print ("Since p_value < 0.05, we reject H0 with a confidence of 95%")

#To find the confidence level
print ("We can reject H0 with a confidence of:", (1-p_val)*100, "%")

print ("The following data is about the correlation between the 2 variables used.")
print ("Correlation data: T empirical: 0.8099, and the P_VALUE is 0.00")
#To do a scatter plot
plt.scatter(my_data.gdp_pc, my_data.happ_score)
plt.title("FIGURE 11. Scatter Plot GDP per Capita and Happ Score")
plt.xlabel("GDP pc\nSource: Happiness Ranking.2017-csv data")
textstr= "T emp=0.8099\nP_value=0.00\nn=153"
plt.text(1.40,3.55,textstr)
# happiness index by life expectancy
stats.pearsonr(my_data.gdp_pc, my_data.happ_score)

#pearson correlation coefficient= T empirical  , p value
```
